```python
# لیست کلمات با خط تیره
word_list = ["ناحیه-1", "منطقه-2", "شهر-3"]

# اضافه کردن "تهران" به انتهای هر کلمه با خط تیره
neighborhoods2 = [word + " تهران" for word in adress]

print(len(neighborhoods2))
```

```python
from geopy.geocoders import Nominatim
import pandas as pd

# تعریف ناحیه‌های مورد نظر
neighborhoods = ["نیاوران، تهران", "شهرک غرب، تهران", "منطقه ۲۲ تهران", "ونک، تهران"]

# ایجاد یک لیست برای نگهداری اطلاعات جغرافیایی هر ناحیه
geo_data = []


# اتصال به سرویس مکان‌یابی
geolocator = Nominatim(user_agent="geocoder")

# ایجاد یک batch size
batch_size = 2

# تقسیم نواحی به دسته‌هایی با batch_size تعدادی
batches = [neighborhoods2[i:i + batch_size] for i in range(0, len(neighborhoods2), batch_size)]

# دریافت اطلاعات جغرافیایی برای هر دسته از نواحی
for batch in batches:
    for neighborhood in batch:
        location = geolocator.geocode(neighborhood,timeout=10)
        if location:
            geo_data.append({
                "ناحیه": neighborhood,
                "طول جغرافیایی": location.longitude,
                "عرض جغرافیایی": location.latitude
            })

# تبدیل لیست به یک فریم داده (DataFrame) با کمک کتابخانه pandas
df = pd.DataFrame(geo_data)

# ذخیره فریم داده به یک فایل CSV
df.to_csv("data3.csv", index=False, encoding="utf-8")

print("اطلاعات جغرافیایی ذخیره شد.")
```

```python
def remove_direction_words(text):
    direction_words = ["شمالی", "جنوبی", "شرقی", "غربی"]
    for word in direction_words:
        text = text.replace(word, "")
    return text.strip()

# تابع برای حذف کلمات قبل از "-" و تمیز کردن داده
def clean_data(data):
    cleaned_data = []
    for item in data:
        if "-" in item:
            parts = item.split("-")
            cleaned_text = remove_direction_words(parts[-1])  # حذف کلمات شمالی، جنوبی، شرقی، غربی
            cleaned_data.append(cleaned_text)
        else:
            cleaned_text = remove_direction_words(item)  # حذف کلمات شمالی، جنوبی، شرقی، غربی
            cleaned_data.append(cleaned_text)
    return cleaned_data

# لیست اصلی داده‌ها
data = ["بلوار فردوس شرقی، تهران", "بلوار فردوس غربی، تهران", "خیابان شمالی مثلث، تهران",
        "خیابان جنوبی مثلث، تهران", "جاده شرقی شهرک", "خیابان مرکزی غربی"]

# پاک کردن داده‌ها و ذخیره در لیست جدید
cleaned_data = clean_data(missing_data)

len(cleaned_data)
```
